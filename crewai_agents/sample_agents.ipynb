```python
from dotenv import load_dotenv
load_dotenv()
```

```python

from crewai import LLM

llm = LLM(
    model="gemini-2.5-flash",
    temperature=0.1
)
llm.call("Who invented transcendental meditation.")

```

```python
from crewai import Agent, Task, Crew

whatsapp_reader = Agent(
    role = "whatsapp message reading assistant",
    goal = "to take the message as input and give the exact message in a more understandable format like expand all the typical whatsapp terms to general english",
    backstory = "a youngster who can decode all the whatsapp jargons and tell the exact message",
    verbose = True,
    llm = llm
)

message = "btw, hru"

decode_task = Task(
    description = f"take the following message and remove all the jargons and replace them with actual sentences, message: '''{message}''' ",

    agent = whatsapp_reader,

    expected_output= "A well understandable message without any confusing words or jargons"


)

crew = Crew(
    agents = [whatsapp_reader],
    tasks=[decode_task],
    verbose = True
)

crew.kickoff()
```
